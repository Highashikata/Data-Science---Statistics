```python
billboard_short = df[['artist.inverted','track', 'time', 'date.entered', 'x1st.week', 'x2nd.week', 'x3rd.week', 'x4th.week', 'x5th.week']]

billboard_short.head()
```

```python
billboard_short.columns = ['Artist', 'Track', 'Time', 'Entering Date', 'Week1', 'Week2', 'Week3', 'Week4', 'Week5']
```

```python
billboard_short.head()
```

```python
billboard_short
```

```python
melt_the_week_data = pd.melt(billboard_short, id_vars=['Artist', 'Time', 'Entering Date']) 

melt_the_week_data.head()

melt_the_week_data['variable'].unique()
```

Then, we will use the melt function in order to pivot the 3 columns (week1,2,3) into one colulmn named 'Week' and for the value of the every week we will create another column named rank.

```python
melt_df = pd.melt(billboard_short, id_vars = ['Artist', 'Time', 'Entering Date'], var_name = 'Week', value_name = 'Rank')
```
